```python
from Bio.PDB import *
from biopandas.pdb import PandasPdb
import numpy as np
import sys
```

```python
new = PandasPdb().read_pdb('6nb8_imgt.pdb')
# new.head()
# #new1 = new.df['ATOM'].head()
kinley = new.df['ATOM']
draft = kinley.drop(['blank_3', 'blank_2','blank_1', 'alt_loc','insertion','blank_4','segment_id'],axis=1)
draft
```

```python
#  H data
H_chain = draft[draft['chain_id'] == 'H']
H_chain = H_chain[H_chain['residue_number'] <= 126]
H_chain = H_chain[H_chain['residue_number'] >=1]
H_chain
```

```python
#  L data
L_chain = draft[draft['chain_id'] == 'L']
L_chain = L_chain[L_chain['residue_number'] <= 127]
L_chain = L_chain[L_chain['residue_number'] >= 1]
L_chain
```

```python
import math

def e_dist(x1, y1, z1, x2, y2, z2):
    return math.sqrt((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)


valid_bonds = {}

num = 0
for _, H_row in H_chain.iterrows():
    
    x_1, y_1, z_1 = H_row['x_coord'], H_row['y_coord'], H_row['z_coord']
    
    for _, L_row in L_chain.iterrows():
        x_2, y_2, z_2 = L_row['x_coord'], L_row['y_coord'], L_row['z_coord']
        
        distance = e_dist(x_1,y_1,z_1,x_2,y_2,z_2)
        
        if distance <= 5:
           # print('Bond found!' + 'Distance = ', distance)
            num += 1
            
            valid_bonds[str(num)] = {'chain' : [H_row['atom_number'],  L_row['atom_number']],
                                    'dist': distance}
            A = H_row['residue_number']
            B = L_row['residue_number']
            print(A,B)
#         else:
#             print("Bond can't be formed!"+ 'Distance = ', distance)
            
        
        
        
  # print('Bond length between the atoms with serial number {} and {} is {:.5f}'.format(coords[i][1], coords[j][1], dist))     
        
```
